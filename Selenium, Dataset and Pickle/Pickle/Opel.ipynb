```python
def data_manipulation():
    for col in seri_columns:
        arac_sayisi = len(opel_Df.groupby(col).get_group(1))
        index = opel_Df.groupby(col).get_group(1).index
        if arac_sayisi < 100:
            opel_Df.drop(index,axis=0,inplace=True)
            opel_Df.drop(col,axis=1,inplace=True)
```

```python
data_manipulation()
```

# Makine Öğrenmesi

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = opel_Df.drop("Fiyat",axis=1)
y = opel_Df["Fiyat"]
```

```python
train_x, test_x ,train_y, test_y = train_test_split(x,y,test_size=0.3,random_state=33)
```

```python
model = LinearRegression()
```

```python
model.fit(train_x,train_y)
```

```python
predict = model.predict(test_x)
```

```python
from sklearn.metrics import r2_score
r2_score(test_y, predict)
```

```python
yeni = opel_Df.loc[test_y.head(50).index].Fiyat
yeni_df = pd.DataFrame(data=yeni.values)
```

```python
yeni_df.rename(columns={0:"Gerçek Fiyat"},inplace=True)
```

```python
yeni_df["Tahmin"] = predict[0:50]
```
