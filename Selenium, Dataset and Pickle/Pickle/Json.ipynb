```python
data = pd.read_excel('Arabalar.xlsx')
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
data
```

```python
markalar = data.Marka.unique()
```

```python
data[data["Marka"] == "Alfa Romeo"].Seri.unique()
```

```python
data[(data["Marka"] == "Audi") & (data["Seri"] == "A3")].Model.unique()
```

```python
def json_load():
    string = ""
    for marka in markalar:
        dict = {
        "id": marka,
        "name": marka,
        "parent_id": "0"
        }
        string += str(dict) + ","
        seriler = data[data["Marka"] == marka].Seri.unique()
        for seri in seriler:
            dict = {
            "id": seri,
            "name": seri,
            "parent_id": marka
            }
            string += str(dict) + ","
            models = data[(data["Marka"] == marka) & (data["Seri"] == seri)].Model.unique()
            for model in models:
                dict = {
                "id": model,
                "name": model,
                "parent_id": seri
                }
                string += str(dict) + ","
    return string
```
