```python
df.loc[df["Boya-değişen"] == "Belirtilmemiş", "Boya"] = 0
df.loc[df["Boya-değişen"] == "Belirtilmemiş", "Değişen"] = 0
df.loc[df["Boya-değişen"] == "Tamamı orjinal", "Boya"] = 0
df.loc[df["Boya-değişen"] == "Tamamı orjinal", "Değişen"] = 0
df.loc[df["Boya-değişen"] == "Tamamı boyalı", "Boya"] = 12
df.loc[df["Boya-değişen"] == "Tamamı boyalı", "Değişen"] = 0
```

```python
def change_boya_degisen(obj):
    if ("değişen" in obj) and ("boyalı" in obj):
        return obj.replace(" değişen, ","-").replace(" boyalı","")
    else: return obj
df["Boya-değişen"] = df["Boya-değişen"].apply(change_boya_degisen)
```
